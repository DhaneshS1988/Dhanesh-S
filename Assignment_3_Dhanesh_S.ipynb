```python
#doing hyper parameter tuning
```

```python
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_classes=2, random_state=42)
```

```python
from scipy.stats import randint
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    "max_depth": [3, None],
    "max_features": randint(1, 9),
    "min_samples_leaf": randint(1, 9),
    "criterion": ["gini", "entropy"]
}
```

```python
tree = DecisionTreeClassifier()
tree_cv = RandomizedSearchCV(tree, param_dist, cv=5)
tree_cv.fit(x, y)

print("Tuned Decision Tree Parameters: {}".format(tree_cv.best_params_))
print("Best score is {}".format(tree_cv.best_score_))
```

```python
#best score obtained here is 0.824
```

```python
#grid search
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

from sklearn.datasets import make_classification
```

```python
x, y = make_classification(
    n_samples=1000, n_features=20, n_informative=10, n_classes=2, random_state=42)
```

```python
#creating hyperparemeter grid
```

```python
c_space = np.logspace(-5, 8, 15)
param_grid = {'C': c_space}
```

```python
#doing logistic regression now
```

```python
logreg = LogisticRegression()
```

```python
logreg_cv = GridSearchCV(logreg, param_grid, cv=5)
```

```python
logreg_cv.fit(x, y)
```

```python
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
print("Best score is {}".format(logreg_cv.best_score_))
```

```python
#here the score obtained is 0.853
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint as sp_randint
```

```python
clf = RandomForestClassifier(n_estimators=50)
```

```python
param_dist = {"max_depth": [3, None],
              "max_features": sp_randint(1, 11),
              "min_samples_split": sp_randint(2, 11),
              "min_samples_leaf": sp_randint(1, 11),
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}
```

```python
samples=8
randomCV = RandomizedSearchCV(clf, param_distributions=param_dist, n_iter=samples,cv=3)
```

```python
randomCV.fit(x, y)
print(randomCV.best_params_)
```

```python
randomCV.cv_results_['mean_test_score']
```
