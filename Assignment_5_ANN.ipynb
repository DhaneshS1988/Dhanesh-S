```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
model=Sequential()
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
def create_model(optimizer='SGD'):
#create a model
  model=Sequential()
  model.add(Dense(12,input_shape=(20,),activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
  return model
```

```python
seed=7
tf.random.set_seed(seed)
```

```python
model=KerasClassifier(model=create_model,batch_size=10,verbose=0)
```

```python
optimizer = ['SGD',  'Adam', 'Adamax', 'Nadam']
epochs = [10,20,30]
learn_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
param_grid = dict(model__optimizer=optimizer,optimizer__learning_rate=learn_rate,epochs=epochs)
```

```python
grid=GridSearchCV(estimator=model,param_grid=param_grid,n_jobs=-1,cv=3)
grid_result=grid.fit(x,y)
```

```python
#summarize the results
print("Best: %f using %s" % (grid_result.best_score_,grid_result.best_params_))
meanscore = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(meanscore, stds, params):
  print("mean= ", mean, "stdev= ",stdev,'with',param)
```

```python
#the best value obtained here is
#Best: 0.250000 using {'epochs': 10, 'model__optimizer': 'SGD', 'optimizer__learning_rate': 0.001}
```

```python
from tensorflow.keras.regularizers import l2
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
def build_model(lambda_parameter=0.1):
    model = Sequential()
    model.add(Dense(8, input_dim = 8, activation = "relu",
                  kernel_regularizer = l2(lambda_parameter)))
    model.add(Dense(8, activation = "relu",
                  kernel_regularizer = l2(lambda_parameter)))
    model.add(Dense(1, activation = "sigmoid"))
    model.compile(loss = "binary_crossentropy", optimizer = "sgd",
                metrics = ["accuracy"])

    return model
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
seed = 1
np.random.seed(seed)
```
